```python
import pandas as pd
left = pd.DataFrame({
    'id':[1,2,3,4,5],
    'Name':['A','B','C','D','E'],
    'category':['cat1','cat2','cat3','cat4','cat5']
})
right = pd.DataFrame({
    'id':[1,2,3,4,5],
    'Name':['F','G','H','I','J'],
    'category':['cat2','cat4','cat3','cat6','cat5']
})
```

```python
left
```

```python
right
```

```python
pd.merge(left, 
         right, 
         how = 'inner',
         on = 'category',
         left_on = None,
         right_on = None, 
         left_index = False,
         right_index = False,
         sort = True
)
```
