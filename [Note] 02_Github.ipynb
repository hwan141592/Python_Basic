# Github

Git vs. GitHub

- Git
    - 개발을 진행하며 작성하는 소스코드가 업데이트되는 버전을 기록해두고 관리할 수 있는 소스코드 버전 관리 시스템
- GitHub
    - Git으로 관리하는 프로젝트를 호스팅하고 시간과 공간의 제약 없이 협업할 수 있는 온라인 서비스
- Git 이 버전 기록을 저장한다면, GitHub에서는 그 기록을 다른 사람과 함께 공유하며 협업 가능.
- Local에서 작업한 내용을 Git이 저장해 두었다면, 그 기록을 온라인 작업공간인 GitHub에 올려 원격(Remote)으로도 작업할 수 있도록 한다.

## Git 초기 셋팅

### [1] Local의 Git에 Github 계정정보 등록하기

```bash
git config --global user.email "my_email@gmail.com"
git config --global user.name "my_username"
git config -l
```

### [2] 내 컴퓨터에 Local Repository 만들기

```bash
cd ~
mkdir workplace
cd workplace
pwd
git init
```

### [3] README.md

```bash
$ echo "# first-repository" >> README.md
$ cat README.md
$ git status
$ git log
```

[Git 저장소 생성 및 커밋 ( init / add / commit )](https://ifuwanna.tistory.com/193)

### [4] git add/commit/push

```bash
$ git add .
$ git commit -m 'readme commit'
$ git remote add origin https://github.com/user_name/repository_name
$ git push origin master
```

### [5] 기타

```bash
$ git clone https://... #github repository url
$ echo "add and correct cloned files" >> README.md
$ git status
$ git add .
$ git commit -m 'modify commit'
$ git push origin master

$ git pull origin master   # when remote repository was modified
```

### Git Repository 구조

- 작업폴더(Working Directory) >> Index(Staging Area) >> 저장소(Head-Repository)
- 저장소에 commit하기 위해서 먼저 추가(Untracked Files) 및 변경(Modified files)하고자 하는 파일을 먼저 인덱스에 기록(Stage)하고 이후 스테이징된 목록만 최종적으로 commit 명령어에 의해 저장소에 공개.
- index에 추가된 파일을 제외하려면 >> git rm —cached 명령어 사용.
- git commit -a  (옵션 -a) >>> 스테이징 절차(add)를 생략하고 바로 add와 commit 동시에 가능.
